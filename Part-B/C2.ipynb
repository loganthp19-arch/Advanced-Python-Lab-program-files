```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv('Automobile_data.csv', na_values=['?'])
df.head()
```

```python
# i) Plot for number of doors by fuel type
sns.countplot(data=df[df['num-of-doors'] != '?'], x='num-of-doors', hue='fuel-type')
plt.title('Two and Four Door Cars by Fuel Type')
plt.show()

# ii) Plot for body style by fuel type
sns.countplot(data=df, x='body-style', hue='fuel-type')
plt.title('Car Body Styles by Fuel Type')
plt.show()

# iii) Boxplot for horsepower by fuel type and drive wheels
# Convert horsepower to float after filtering out '?'
df_hp = df[df['horsepower'] != '?'].copy()
df_hp['horsepower'] = df_hp['horsepower'].astype(float)

body_style_counts = df.groupby(['body-style', 'fuel-type']).size().reset_index(name='count')
print("Total number of cars by body style and fuel type:\n", body_style_counts)

sns.boxplot(data=df_hp, x='drive-wheels', y='horsepower', hue='fuel-type')
plt.title('Horsepower by Fuel Type and Drive Wheels')
plt.show()
```
