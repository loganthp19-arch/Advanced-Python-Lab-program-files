# Box Plot Interpretation 

```python
import pandas as pd
import seaborn as sns

auto = pd.read_csv('Automobile_data.csv',na_values=['?'])
```

```python
auto.head()
```

```python
auto.describe()
```

```python
#distribution of the the two and four door cars with respect to the type of fuel they use. 

sns.countplot(x= auto['num-of-doors'],hue= auto['fuel-type']);

#it seems there are fewer four door diesel cars that gasoline cars  and  same stands for the two door cars as well.
```

```python
#Distribution of cars of different body styles with respect to the type of fuel they use

sns.histplot(x = auto['body-style'], hue=auto['fuel-type']);

# it can be seen  for some body-style the diesel car doesnt even exist(convertible)
```

```python
#Total number of each type of body style cars categorized by fuel type
sns.countplot(x = auto['body-style'], hue= auto['fuel-type']);

#it can be observed that sedan cars of gas fuel type are more. There are no diesel convertible cars
```

```python
#Horsepower of each of the fuel type with reference to the type of drive wheel present in cars

sns.catplot(x="fuel-type",
            y= "horsepower",
            hue = "num-of-doors",
            col= "drive-wheels",
            data = auto,
            kind = "box"
           );

# with reference to the type of drive wheel the car is. From data we can see there are many real wheel drive cars andseems to have the highest 
# horsepower as well in gasoline category under two or four door cars.
```

```python
#or can plot box plot, but insights obtained is less compared to categorical plot
sns.boxplot(x = auto['fuel-type'], y = auto['horsepower'], hue=auto['drive-wheels']);
```
